Vamos trabalhar com mais de um DataFrame

```python
import pandas as pd

filePath='D:\\dev\\github.com\\ricardozanini\\data-science-course\\aula4\\vendas.xlsx'
```

```python
vendas1=pd.read_excel(filePath,sheet_name='vendas_sede')
vendas1
```

```python
vendas2=pd.read_excel(filePath,sheet_name='vendas_filial')
vendas2
```

```python
clientes=pd.read_excel(filePath,sheet_name='clientes')
clientes.head()
```

```python
produtos=pd.read_excel(filePath,sheet_name='produtos')
produtos.head()
```

```python
#Vamos gerar o dataframe vendas concatenando vendas1 e vendas 2 pelas linhas (axis=0)
vendas=pd.concat([vendas1,vendas2],axis=0,ignore_index=True)
vendas
```

```python
#vamos fazer um merge (inner) entre vendas e produtos para apresentar os detalhes dos produtos
detalhes=vendas.merge(right=produtos,how='inner',left_on='id_produto',right_on='sku')
detalhes.head()
```

```python
#usando as vendas do usuário id=2, faça novo inner com clientes

detalhes2=detalhes[detalhes['id_cliente']==2].merge(right=clientes, how='inner', left_on='id_cliente', right_on='id')
detalhes2
```

```python
#Desafio: usando conjuntos, encontre os usuários que compram na sede E na filial
set(vendas1['id_cliente'])&set(vendas2['id_cliente'])
```

```python
#Faça um outer entre vendas1 e vendas 2, use indicator=True
outer=vendas1.merge(right=vendas2,how='outer',on='id_cliente',indicator=True)
outer
```

```python
#Agora, um left , assim todas aschaves da esquerda entram...

left=vendas1.merge(right=vendas2,how='left',on='id_cliente',indicator=True)
left
```
