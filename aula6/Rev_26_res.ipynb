Um pouco mais de Gradient Descent

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

Minimizar a função f(x,y)=cos(2x)+y**2+y

```python
# Obtendo f e o gradiente de f
def f(x,y):
    return math.cos(2*x)+y**2+y
def grad(x,y):
    return (-2*math.sin(2*x),2*y+1)
```

```python
#Fazendo Gradient Descent, traçando as curvas de nível e o caminho da minimização

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z = np.cos(2*X)+Y**2+Y
fig, ax = plt.subplots(figsize=(10,10))

CS = ax.contour(X, Y, Z,levels=20)
ax.clabel(CS, inline=1, fontsize=8)
ax.set_title('Simplest default with labels')
x,y=-1.0,-2.0
lr=0.1
ax.plot(x,y,'bo')
for i in range(2000):
    
    g=grad(x,y)
    x=x-lr*g[0]
    y=y-lr*g[1]
    ax.plot(x,y,'bo')

plt.show()
```
