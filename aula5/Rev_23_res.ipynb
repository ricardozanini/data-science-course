Autovalores e autovetores

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Observe a transformação do vetor v ao ser multiplicado pelas matrizes M1 e M2 
v=np.array([0.5,0.5])
plt.arrow(0, 0, v[0], v[1], head_width=0.05, head_length=0.05, fc='k', ec='k')
plt.show()
```

```python
#Transformação 1
M1=np.array([[0.3,0.3],[-0.1,0.3]])
vl=np.dot(M1,v)
print(vl)
plt.title('Tranformação 1 ')
plt.arrow(0, 0, vl[0],vl[1], head_width=0.05, head_length=0.05, fc='k', ec='k')
plt.show()
```

```python
#Transformação 2
M2=np.array([[0.2,0.5],[0.3,0.4]])
vl=np.dot(M2,v)
print(vl)
plt.title('Tranformação 2 ')
plt.arrow(0, 0, vl[0],vl[1], head_width=0.05, head_length=0.05, fc='k', ec='k')
plt.show()
```

```python
# A transformação 2 foi equivalente a multiplicar v pelo escalar 0.7
#Transformação 2
M2=np.array([[0.2,0.5],[0.3,0.4]])
vl=0.7*v
print(vl)
plt.title('Multiplicação de v por escalar 0.7 ')
plt.arrow(0, 0, vl[0],vl[1], head_width=0.05, head_length=0.05, fc='k', ec='k')
plt.show()
```

```python
#Quando multiplicamos uma matriz M por um vetor v e isso tem o mesmo efeito de multiplicar v por lambda
#Dizemos que v é um autovetor de e lambda é o autovalor associado
#Assim, M2 tem autovetor v e 0.7 é o autovalor associado
#Vamos partir de M e chegar a v e lambda com numpy
a_val,a_vet=np.linalg.eig(M2)
a_val,a_vet
#Cada coluna de a_vet é um autovetor unitário que tem como autovalor associado o elemento correspondente de a_val


```

```python
#Vamos testar os dois pares de autovalor e autovetor de M2..primeiro
np.dot(M2,a_vet[:,0]),a_val[0]*a_vet[:,0]
```

```python
#Vamos testar os dois pares de autovalor e autovetor de M2..segundo
np.dot(M2,a_vet[:,1]),a_val[1]*a_vet[:,1]
```
